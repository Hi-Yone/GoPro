# 加速度・角速度データのプロット

```python
# https://gopro.github.io/gpmf-parser/
# data order : [Z, X, Y]

def extract_IMUdata_fromJson(path):
    file = open(path, 'rb').read()
    data = json.loads(file.decode('utf-8'))

    AcclData = data['1']['streams']['ACCL']['samples']
    GyroData = data['1']['streams']['GYRO']['samples']

    dataSize = len(AcclData)
    accl = np.zeros((dataSize,3))
    gyro = np.zeros((dataSize,3))
    for ii in range(dataSize):
        accl[ii] = (AcclData[ii]['value'])
        gyro[ii] = (GyroData[ii]['value'])

    return [accl, gyro]

def plot_Accldata(sensordata=list, direction=str, ifig=int):
    accl, gyro = sensordata
    xx = np.arange(len(accl))
    
    plt.subplot(3,3,ifig); ifig+=3
    plt.title(f'{direction} \n Z')
    plt.plot(xx, accl[:,0])
    plt.yticks(np.arange(-10, 30.1, 10))
    plt.ylim(-10, 32)

    plt.subplot(3,3,ifig); ifig+=3
    plt.title('X')
    plt.plot(xx, accl[:,1])
    plt.yticks(np.arange(-10, 30.1, 10))
    plt.ylim(-10, 32)

    plt.subplot(3,3,ifig)
    plt.title('Y')
    plt.plot(xx, accl[:,2])
    plt.yticks(np.arange(-10, 30.1, 10))
    plt.ylim(-10, 32)

def plot_Gyrodata(sensordata=list, direction=str, ifig=int):
    accl, gyro = sensordata
    xx = np.arange(len(gyro))
    
    plt.subplot(3,3,ifig); ifig+=3
    plt.title(f'{direction} \n Z')
    plt.plot(xx, gyro[:,0])
    plt.yticks(np.arange(-2, 2.1, 2))
    plt.ylim(-2, 2)

    plt.subplot(3,3,ifig); ifig+=3
    plt.title('X')
    plt.plot(xx, gyro[:,1])
    plt.yticks(np.arange(-2, 2.1, 2))
    plt.ylim(-2, 2)

    plt.subplot(3,3,ifig)
    plt.title('Y')
    plt.plot(xx, gyro[:,2])
    plt.yticks(np.arange(-2, 2.1, 2))
    plt.ylim(-2, 2)
```

## センサのキャリブレーション

```python
# キャリブレーション関数
def calibrate_sensor(num_samples, accl_avg, gyro_avg):
    accl_bias = np.zeros(3)  # 加速度のバイアス
    gyro_bias = np.zeros(3)   # 角速度のバイアス
    accl_scale_factor = np.ones(3)  # 加速度のスケールファクター
    gyro_scale_factor = np.ones(3)   # 角速度のスケールファクター

    # 指定されたサンプル数のデータを収集して平均を計算する
    for nn in range(num_samples):
        accl_data = accl_avg[nn]
        gyro_data = gyro_avg[nn]

        # バイアスの更新
        accl_bias += accl_data
        gyro_bias += gyro_data

    # 平均を取る
    accl_bias /= num_samples
    gyro_bias /= num_samples

    # 推定されたバイアスを表示
    print("Estimated Acclerometer Bias:", accl_bias)
    print("Estimated Gyroscope Bias:", gyro_bias)

    # センサからのデータを[-1, 1]の範囲に正規化する
    max_val = np.max(np.abs(accl_data))
    accl_scale_factor = max_val / accl_bias

    max_val = np.max(np.abs(gyro_data))
    gyro_scale_factor = max_val / gyro_bias

    # 推定されたスケールファクターを表示
    print("Estimated Acclerometer Scale Factor:", accl_scale_factor)
    print("Estimated Gyroscope Scale Factor:", gyro_scale_factor)

    # バイアスとスケールファクターを返す
    return accl_bias, gyro_bias, accl_scale_factor, gyro_scale_factor
```

```python
import numpy as np
from scipy.optimize import leastsq
import json

# データ読み込み関数
def extract_IMUdata_fromJson_calibration(path, pose_num=30, frame_num=100):
    accl = np.zeros((pose_num, frame_num, 3))
    gyro = np.zeros((pose_num, frame_num, 3))
    
    for ii in range(pose_num):
        with open(path.format(ii+1), 'rb') as f:
            data = json.loads(f.read().decode('utf-8'))
        
        AcclData = data['1']['streams']['ACCL']['samples']
        GyroData = data['1']['streams']['GYRO']['samples']
        
        for jj in range(frame_num):
            accl[ii, jj] = AcclData[jj]['value']
            gyro[ii, jj] = GyroData[jj]['value']
    
    return accl, gyro

# 加速度キャリブレーションモデル
def calibrate_accel(accl, theta):
    alpha_xy, alpha_xz, alpha_yz, s_x, s_y, s_z, o_x, o_y, o_z = theta
    misalignment = np.array([
        [1, alpha_xy, alpha_xz],
        [0, 1, alpha_yz],
        [0, 0, 1]
    ])
    scale_factors = np.array([s_x, s_y, s_z])
    biases = np.array([o_x, o_y, o_z])
    calibrated_accl = (accl - biases) @ misalignment.T * scale_factors
    return calibrated_accl

# 目的関数の定義
def objective_function(theta, accl):
    errors = []
    for pose in accl:
        for sample in pose:
            calibrated_accl = calibrate_accel(sample, theta)
            norm_squared = np.sum(calibrated_accl ** 2)
            errors.append(1 - norm_squared / 9.81**2)  # 重力加速度を正規化
    return errors

# 初期パラメータの設定
initial_theta = [0, 0, 0, 1, 1, 1, 0, 0, 0]

# データ読み込み
pose_num = 30
frame_num = 100  # 各姿勢でのフレーム数
path = f'/Users/yonekawahiroto/Sanada_lab/研究/gopro_local/outputs/GyroData_{direction}_left.json'
accl, gyro = extract_IMUdata_fromJson_calibration(path, pose_num, frame_num)

# 最適化の実行
result = leastsq(objective_function, initial_theta, args=(accl,))
optimized_theta = result[0]

print("Optimized parameters:", optimized_theta)

# 校正後の加速度データの確認
theta = np.array([[calibrate_accel(sample, optimized_theta) for sample in pose] for pose in accl])
print("Calibrated acceleration data shape:", theta.shape)
```

```python
T = np.array([[1, -optimized_theta[0], optimized_theta[1]],
              [0, 1, -optimized_theta[2]],
              [0, 0, 1]])
S = np.array([[optimized_theta[3], 0, 0],
              [0, optimized_theta[4], 0],
              [0, 0, optimized_theta[5]]])
o = np.array([[optimized_theta[6]],
              [optimized_theta[7]],
              [optimized_theta[8]]])
print(T)
print(S)
print(o)
```
